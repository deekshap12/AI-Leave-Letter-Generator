```python
from google import genai
from getpass import getpass
api_key=getpass("Enter your Gemini API Key:")
client=genai.Client(api_key=api_key)
```

```python
print("AI LEAVE LETTER GENERATOR")
print("-" * 50)

student_name = input("Enter Student Name: ")
class_name = input("Enter Class/Semester: ")
recipient = input("Enter Recipient (Principal / HOD / Teacher): ")

leave_type = input("Enter Leave Type (Sick / Personal / Family Function / Emergency): ")

reason = input("Enter Reason for Leave: ")
from_date = input("Enter Leave Start Date: ")
to_date = input("Enter Leave End Date: ")

tone = input("Enter Tone (Formal / Polite / Professional): ")
```

```python
prompt = f"""
You are an academic assistant.

Write a complete student leave letter using the information below.

Student Name: {student_name}
Class: {class_name}
Recipient: {recipient}
Leave Type: {leave_type}
Reason: {reason}
Leave Duration: {from_date} to {to_date}
Tone: {tone}

Instructions:
1. Write a simple leave letter from the student.
2. Use the student name and class directly.
3. Do NOT include placeholders like [Date], [School Name], or [Parent Name].
4. The letter should be ready to submit without editing.
5. Keep the language polite and professional.
6. Do not write it from the parent's perspective.
"""
```

```python
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=prompt
)

print("-" * 50)
print(response.text)
```
